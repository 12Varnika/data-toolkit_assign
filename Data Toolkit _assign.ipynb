```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset with categories and numerical values
np.random.seed(0)  # Seed for reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data = {
    'Category A': np.random.normal(50, 10, 100),  # 100 values for Category A, mean=50, std=10
    'Category B': np.random.normal(60, 12, 100),  # 100 values for Category B, mean=60, std=12
    'Category C': np.random.normal(55, 8, 100),   # 100 values for Category C, mean=55, std=8
    'Category D': np.random.normal(65, 15, 100)   # 100 values for Category D, mean=65, std=15
}
data_values = [data[category] for category in categories]
plt.figure(figsize=(10, 6))
plt.boxplot(data_values, labels=categories, patch_artist=True, 
            boxprops=dict(facecolor='lightblue'), medianprops=dict(color='red'))
plt.title("Distribution of Numerical Values Across Categories")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#QUES3.  Using the sales data, generate a heatmap to visualize the variation in sales across different months and
#days.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(20)

# Create the sales dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Pivot the data to create a heatmap-friendly format (mean sales for each day and month)
heatmap_data = pd.pivot_table(df, values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt='.0f')
plt.title("Sales Variation Across Different Months and Days")
plt.xlabel("Month")
plt.ylabel("Day")
plt.show()
```

```python
#QUES4.  Using the given x and y data, generate a 3D surface plot to visualize the function.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute z = sin(sqrt(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Convert the data into a DataFrame (optional)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)
ax.set_title("3D Surface Plot of Z = sin(sqrt(x^2 + y^2))")
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.show()
```

```python
#QUES.2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#markers based on the 'sizes' and 'colors' columns.
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis256

# Display the plot in the notebook
output_notebook()

# Generate random x and y values
np.random.seed(42)
x = np.random.rand(50) * 100  # Random values between 0 and 100
y = np.random.rand(50) * 100

# Generate random sizes and colors
sizes = np.random.randint(10, 50, size=50)  # Random sizes between 10 and 50
colors = np.random.rand(50) * 100  # Random color values to map to a color palette

# Create a ColumnDataSource to hold the data
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a scatter plot
p = figure(title="Bokeh Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Define the color mapper (using Viridis color palette)
color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

# Add circle glyphs with size and color variations
p.circle('x', 'y', size='sizes', color=color_mapper, fill_alpha=0.6, source=source)
show(p)
```

```python
#QUES3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Display the plot in the notebook
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a bar chart
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

# Add bars (vbar: vertical bars)
p.vbar(x='fruits', top='counts', width=0.6, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Customize plot
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

```python
#QUES4.  Create a Bokeh histogram to visualize the distribution of the given data.
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display the output in the notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Calculate the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure
p = figure(title="Histogram of Data", background_fill_color="#fafafa")

# Plot the histogram using 'quad'
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="navy", line_color="white", alpha=0.7)

# Add labels
p.yaxis.axis_label = 'Frequency'
p.xaxis.axis_label = 'Value'

# Show the plot
show(p)
```

```python
#QUES.5 Create a Bokeh heatmap using the provided dataset.
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Display the output in the notebook
output_notebook()

# Generate random heatmap data
data_heatmap = np.random.rand(10, 10)

# Create x and y coordinates
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', 
           x_range=(0, 1), y_range=(0, 1), toolbar_location=None)

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Add the heatmap image
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)

# Add grid lines and customize the layout
p.grid.grid_line_color = None
p.xaxis.axis_label_standoff = 12
p.yaxis.axis_label_standoff = 12

# Show the plot
show(p)
```
